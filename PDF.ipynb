Rolling two dice 10000 times and plotting PDF


```python
import random
l=[]
for i in range(10000):
   a= random.randint(1,6)

   b=random.randint(1,6)

   l.append(a+b)
```

```python
l[:6]
```

```python
import pandas as pd
pd.Series(l).value_counts().sort_index()
```

```python
s=((pd.Series(l).value_counts())/pd.Series(l).value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind='bar')
```

Plotting CDF for above example

```python
import numpy as np
```

```python
np.cumsum(s)
```

```python
np.cumsum(s).plot(kind='bar')
```

Parametric Density Estimation

```python
import matplotlib.pyplot as plt 
import numpy as np
from  numpy.random import normal

sample=normal(loc=50,scale=5,size=1000)
```

```python
sample
```

```python
sample.mean()
```

```python
#plot histogram to understand the distribution
plt.hist(sample,bins=10)
```

```python
#calculate sample mean and standard deviation
mean=sample.mean()
std=sample.std()
```

```python
#fit the distribution as above paramters
```

```python
from scipy.stats import norm
dist=norm(mean,std)
```

```python
values=np.linspace(sample.min(),sample.max(),100)
```

```python
probabilities=[dist.pdf(value) for value in values]
```

```python
#plot the histogram and pdf
plt.hist(sample,bins=10,density=True)
plt.plot(values,probabilities)
```

```python
#draw PDF using Seaborn
import seaborn as sns
```

```python
sns.distplot(sample)
```

#non Parametric Density Estimation using KDE (Kernel Density Estimation)

```python
#Generate Data
```

```python
sample1=normal(loc=20,scale=5,size=300)
sample2=normal(loc=40,scale=5,size=500)
```

```python
sample=np.hstack((sample1,sample2))
```

```python
sample
```

```python
#plot histogram with bins=50
plt.hist(sample,bins=50)
```

```python
from sklearn.neighbors import KernelDensity
model=KernelDensity(bandwidth=3,kernel='gaussian')
```

```python
#convert data to 2D array
sample=sample.reshape((len(sample),1))
```

```python
model.fit(sample)
```

```python
values=np.linspace(sample.min(),sample.max(),100)
values=values.reshape((len(values),1))
```

```python
probabilities=model.score_samples(values)
probabilities=np.exp(probabilities)
```

```python
plt.hist(sample,bins=50,density=True)
plt.plot(values[:],probabilities)
plt.show()
```

```python
#using Seaborn
sns.kdeplot(sample)
```
